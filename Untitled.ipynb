```python
#please be aware of the version conflicts between openCV and panadas and tensorflow
#using a conda env is tricky but necessary

#i don't have a solution for the version conflicts but after trying conda and pip
#it only worked after i downloaded the cv2 package from conda and everything else from pip or the opposite
# i really don't remember

import pandas as pd
import numpy as np
from tensorflow import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from tensorflow.keras.utils import to_categorical,image_dataset_from_directory
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf

```

```python
import cv2
```

```python
#this is a pipeline for preprocessing the data
#see tensorflow data api for more info but u probably won't need to do more research on that because 
#everything could be done with pandas just shuffle -> resize -> one hot encode (search about it) -> devide by 255 because 
# neaural networks work better with values between 0 and 1

data = image_dataset_from_directory("GarbageData")

print(data.class_names)
classes=data.class_names
data = data.shuffle(700)
data = data.map(lambda x, y: (x/255.0, one_hot(y, depth=6)))

data_itr = data.as_numpy_iterator()
batch = data_itr.next()

train_size = int(len(data)*0.7)
test_size = int(len(data)*0.1)+1
val_size = int(len(data)*0.2)+1

train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)

print(train_size,test_size,val_size)

```

```python
#example training data the 1 represent the class of the object see the list above to know which class is it
fig, ax = plt.subplots(ncols=6, figsize=(20,20))
for idx, img in enumerate(batch[0][:6]):
    ax[idx].imshow(img)
    ax[idx].set_title(str(batch[1][idx]))
```

```python
#this network architecture is almost copied from a video at https://www.youtube.com/watch?v=jztwpsIzEGc
#but you do need to understand the theory behind it to know what you are doing

#in short
#conv2d is a convolutional layer that is used to extract features from images
#maxpooling is used to save your memory by making the images smaller and more pixelated
#flatten is used to convert the 2d array into a 1d array
#dense is a normal neural network layer
#relu is a activation function that is used to make the network non linear 

#relu has a common problem called dead neurons which is when the output of the relu function is always zero other functions
#could perform better but the training is resource expensive so i didn't try around

#softmax: is used with classification problems to make the output of the network a probability distribution
#categorical_crossentropy is a loss function that is used with classification problems
#adam is an optimizer that decides the learning rate for you and optimizes the network weights

model=Sequential()
model.add(Conv2D(25,kernel_size=3, activation='relu', input_shape=(256,256,3),padding="same"))
model.add(MaxPooling2D())
model.add(Conv2D(100,kernel_size=3, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(25 ,kernel_size=3, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(6, activation='softmax'))
model = tf.keras.models.load_model("model2.h5")
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
#model.fit(train, epochs=19,validation_data=val,callbacks=[tf.keras.callbacks.EarlyStopping(patience=2)])
#model.save("model2.h5")
```

```python
img = cv2.imread("photo.jpg")
img = cv2.resize(img, (256,256))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
pred = model.predict(np.array([img/255.0]))
print("Predicted: ", pred)
print("max:",max(pred[0]))
index = np.argmax(pred[0])
print("Predicted class: ", classes[index])
#the max value position represents the class from this list
#['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']
#in the current output 0.882 is the 1st and max probability so the class is cardboard
```
